##Column Standardization and dummy variables

##### **Numerical Features Column Standardization**

```python
scaler = StandardScaler()
#data_num=data_num.values.reshape(1,-1)

for i in col_num:
  data_num[i]=scaler.fit_transform(data_num[[i]])
```

```python
data_num.head()
```

##### **Categorical dummy variables**

```python
data_cat=data.select_dtypes(include="object")
data_cat=data_cat.drop(columns=["Unnamed: 0",'DOL'])
data_cat.head(2)
```

```python
cat_col=data_cat.columns
```

```python

for i in cat_col:
  data_cat[i]=pd.get_dummies(data_cat[i].astype(str))
```

```python
data_cat.head()
```

### Column Standardization and dummy variables

```python
data_stand=pd.concat([data_num, data_cat],axis=1)
data_stand.head()
```

## Chi sqaured test

```python
from scipy.stats import chi2_contingency
```

```python
X=data_stand[['Gender','Specialization']]
```

```python
X=pd.crosstab(data_stand['Gender'],data_stand['Specialization'])
```

```python
stat, p, dof, expected = chi2_contingency(X)
  

alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

* Above test shows that there is no relationsheep between Gender and Specialization
