###  Analysis of Salary with categorical variable and numerical variable





**Analysis of Salary with Gender**

```python
plt.figure(figsize=(8,5))
sns.factorplot(x='Gender', y='Salary',hue='Degree',kind='bar',data=data)
plt.show()
```

* Female with M.tech/M.E degree has highest salary.
* Male with M.sc(tech) degree has highest salry

**Analysis of Salary with CollegeTier**

```python
plt.figure(figsize=(8,5))
sns.factorplot(x='CollegeTier', y='Salary',hue='Degree',kind='bar',data=data)
plt.show()
```

* Student with college tier 1 and M.tech degree has highest salary
* Student with college tier 2 and B.tech degree has highest salry. 

**Analysis of Salary with Specialization**

```python
plt.figure(figsize = (10,4))
sns.barplot('Specialization', 'Salary', data =data, color = 'r', label= 'Specialization')
plt.xticks(rotation=90)
plt.legend()
```

* Specialization with polymer technology has highest salary
* Computer networking has second highest salary 

**Analysis of Salary with CollegeState**

```python
plt.figure(figsize = (10,4))
sns.barplot('CollegeState', 'Salary', data =data, color = 'r', label= 'CollegeState')
plt.xticks(rotation=90)
plt.legend()
```

* Meghalaya student get highest salary.
* Uninion territory get lowest salary.
